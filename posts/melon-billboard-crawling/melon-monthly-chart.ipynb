```python
def clickChartOptions(prevIndice, curIndice):
    mylogger.logger.debug('-'*50)
    mylogger.logger.debug(f'prevIndice, curIndice | {prevIndice}, {curIndice}')
    for lcnt, idx in enumerate(curIndice):
        mylogger.logger.debug(f'lcnt {lcnt} idx {idx}')
        mylogger.logger.debug(f'prevIndice[idx] / curIndice[idx] | {prevIndice[lcnt]} / {curIndice[lcnt]}')
        
        # skip condition
        if lcnt == 3:
            continue
        if lcnt != 4 and prevIndice[lcnt] == curIndice[lcnt]:
#             mylogger.logger.debug(f'prevIndice[idx] / curIndice[idx] | {prevIndice[idx]} / {curIndice[idx]}')
            mylogger.logger.debug(f'<same>')
            continue
        mylogger.logger.debug(f'<not same>')
        if lcnt <= 1:
            esLi = eChartTabs[lcnt].find_elements_by_css_selector('li')[::-1]
        else:
            esLi = eChartTabs[lcnt].find_elements_by_css_selector('li')
#         mylogger.logger.debug(f'lcnt {lcnt} | text {esLi[idx].text}')
        mylogger.logger.debug(f'len of esLi {esLi[idx].text} {len(esLi)}')
        esLi[idx].click()
        
#         concatChartOptions.append(esLi[idx].text) # for debug
        mylogger.logger.debug(f'btn clicked.')
```

```python
prevIndice = [2,-1,-1,0,-1]
curIndice = [3,0,0,0,0]  # start indice
endIndice = [4,0,11,0,0]
indiceHandler = IndiceHandler()
indiceHandler.maxIndice = [5,10,12,1,1]# 2 is enough to hold each first indice
                                       # and can be modified to bare add operation
                                       # with more than 1.
time_i = time.time()
lcnt = 0
while True:
    clickChartOptions(prevIndice, curIndice)#, maxIndice)#, mode=1)
    if curIndice == endIndice:
        break
    prevIndice = curIndice[:]
    curIndice = indiceHandler.addIndiceToIndice(prevIndice, [0,0,0,0,1])
#     print(f'while - prevIndice curIndice | {prevIndice} {curIndice}')
#     print(f'while - indiceHandler.maxIndice | {indiceHandler.maxIndice}')
    if lcnt%12 == 0:
        print(f'elapsed : {time.time() - time_i}')

    time.sleep(0.25)
    lcnt += 1
print(f'elapsed : {time.time() - time_i}')
```
